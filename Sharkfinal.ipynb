```python
import pandas as pd
import numpy as np
import re
```

```python
#import de dados para iniciar a limpeza de base
data = pd.read_csv ('data/attacks.csv', encoding = "ISO-8859-1")
```

```python
#retirar linhas sem data
data = data[data['Date'].notnull()]

#retirar linhas sem local
data = data[data['Country'].notnull()]

#exibir display completo
pd.set_option ('max_columns',30)
pd.set_option ('max_colwidth',100)

#Drop colunas inuteis no Dataframe
data.drop('Unnamed: 22', axis=1, inplace=True)
data.drop('Unnamed: 23', axis=1, inplace=True)

#Ajuste de substituição de dados 
data['Fatal (Y/N)'].replace('N ','N',inplace=True)
data['Fatal (Y/N)'].replace('Y ','Y',inplace=True)
data['Fatal (Y/N)'].replace('y','Y',inplace=True)
data['Fatal (Y/N)'].replace('n','N',inplace=True)

```

```python
data['Country'].replace('UNITED ARAB EMIRATES (UAE)','UNITED ARAB EMIRATES',inplace=True)
data['Country'].replace('ST HELENA, British overseas territory','UNITED KINGDOM',inplace=True)
data['Country'].replace('ST. MAARTIN','ST. MARTIN',inplace=True)
data['Country'].replace('ENGLAND','UNITED KINGDOM',inplace=True)
data['Country'].replace('NORTH PACIFIC OCEAN','PACIFIC OCEAN',inplace=True)
data['Country'].replace('MID-PACIFC OCEAN','PACIFIC OCEAN',inplace=True)
data['Country'].replace('SOUTH PACIFIC OCEAN','PACIFIC OCEAN',inplace=True)
data['Country'].replace('SOUTHWEST PACIFIC OCEAN','PACIFIC OCEAN',inplace=True)
data['Country'].replace('MID ATLANTIC OCEAN','ATLANTIC OCEAN',inplace=True)
data['Country'].replace('SOUTH ATLANTIC OCEAN','ATLANTIC OCEAN',inplace=True)
data['Country'].replace('NORTH ATLANTIC OCEAN','ATLANTIC OCEAN',inplace=True)
data['Country'].replace('INDIAN OCEAN?','INDIAN OCEAN',inplace=True)
data['Country'].replace('CEYLON (SRI LANKA)','SRI LANKA',inplace=True)
data['Country'] = data['Country'].str.rstrip() 
data['Country'] = data['Country'].str.upper()
```

```python
#tratamento de paises
data['Country'].unique ()
```

```python
# Retirar casa depois do ano da coluna year
data['Year'] = data['Year'].apply(lambda x: str(x).replace('.0',''))
```

```python
#limpando os anos que estão sem informação
data = data.loc[data['Year']!='nan']
```

```python
#tratamento de replaces em linhas
data['Sex '].replace('lli','N',inplace=True)
data['Sex '].replace('M ','M',inplace=True)
data['Name'].replace('?brahim Karagöz','Ibrahim Karagöz', inplace=True)
data['Name'].replace('__ Jimenez','Jimenez', inplace=True)

#Limpando espaço no ínicio dos nomes
data['Name'] = data['Name'].str.lstrip() 
```

```python
#filtrando o nome e alterando dado na coluna sexo
data.loc[data['Name']=='Stephen Pettigew' , 'Sex '] = 'M'
```

```python
data['First'] = data['Name'].apply(lambda x: str(x)[:1])
```

```python
def filtro (x):
    '''função coloca flag nas palavras contidas abaixo'''
    if (re.match('[0-9]|male|female|"|a |an ',x,flags=re.IGNORECASE)):
        return (1)
    else:
        return (0)
```

```python
#transformando a coluna name em string para usar o appy
data['Name']=data['Name'].astype(str)
```

```python
#cria a coluna de flag e aplica o filtro para a limpeza de dados sem nome
data['Flag']= data['Name'].apply(filtro) + data['First'].apply(filtro)
```
